```python
#array to hold the filenames we want
desired_files = []
for file in os.listdir('single_videos'):
    identidy_info = file.split('.')[0]
    identidy_info = identidy_info + '.csv'
    desired_files.append(identidy_info)
    



#print(desired_records)
```

```python
#Grabbing a small dataframe for test purposes
#array to store the relevant frames to merge
frames_to_merge = []
for file in desired_files:
    df_has_file = df_OpenFace[df_OpenFace['Filename']==file] 
    frames_to_merge.append(df_has_file)
    
df_test = pd.concat(frames_to_merge, ignore_index = True)
```

```python
#checking to see if we have any of the newish data
display(df_test)
```
